```python
sky=cv2.imread('Original.jpg')
sky_rgb=np.copy(sky)
sky_rgb=cv2.cvtColor(sky_rgb,cv2.COLOR_BGR2RGB)
plt.imshow(sky_rgb)
```

```python
sky_gray=np.copy(sky)
sky_gray=cv2.cvtColor(sky_gray,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(sky_gray,cmap='gray')
```

```python
sky_gray.shape
```

```python
cv2.minMaxLoc(sky_gray)
```

```python
max_loc_x = cv2.minMaxLoc(sky_gray)[3][0]
max_loc_y= cv2.minMaxLoc(sky_gray)[3][1]
```

```python
s = 200
coord1 = (max_loc_x+s, max_loc_y+s)
coord2 = (max_loc_x-s, max_loc_y-s)
box_color = (300, 0, 0)
border=35

spot_box = np.copy(sky_rgb)
spot_box = cv2.rectangle(spot_box, coord1, coord2, box_color, border)
plt.imshow(spot_box)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 10))
axes[0].imshow(sky_rgb)
axes[0].set_title('Original');
axes[0].axis('off')
axes[1].imshow(spot_box)
axes[1].set_title('Brightest Spot');
axes[1].axis('off')
```
